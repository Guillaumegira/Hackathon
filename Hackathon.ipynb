Calling of the meetup api, with the group url from the given meetup datas. A 3 seconds waiting time is taken, as only a limited number of call per hour is allowed. This API call allows us to get some information about the geo localisation of past event, from the public groups

```python
groups=di["urlname"].unique()
for name in groups:
    print(name)
    url='https://api.meetup.com/'+name+'?&sign=true&photo-host=public&fields=event_sample'
    r=requests.get(url)
    if r.status_code==200:
        data=r.json()
        try: 
            for text in data['event_sample']:
                lon.append(text['venue']['lon'])
                lat.append(text['venue']['lat'])
        except:
            print('pas de sorties')
    else:
        print('groupe inexistant') 
    time.sleep(3)
```
